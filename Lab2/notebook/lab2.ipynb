```python
import sys
import os
from pathlib import Path

ROOT = Path.cwd().resolve()
while ROOT != ROOT.parent and not (ROOT / "data").is_dir():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "data"))
os.chdir(ROOT)
```

```python
from main import (
    run_eda,
    get_cached_split,
    run_bagging,
    run_gradient_boosting,
    run_gbm_comparison,
    run_optuna_tuning,
    run_metrics_demo,
    run_submission,
    TRAIN_PATH,
    SUBMISSION_PATH,
    DEFAULT_TEST_SIZE,
    DEFAULT_RANDOM_STATE,
)

from pathlib import Path
```

```python
run_eda()
```

```python
data_split, split_key = get_cached_split(
    csv_path=TRAIN_PATH,
    test_size=DEFAULT_TEST_SIZE,
    random_state=DEFAULT_RANDOM_STATE,
)
data_split[:2]
```

```python
bagging_scores = run_bagging(data_split, split_key)
grad_scores = run_gradient_boosting(data_split, split_key)
comp_scores, best_model = run_gbm_comparison(data_split, split_key)
best_name, best_params, optuna_score = run_optuna_tuning(
    data_split,
    split_key,
    (comp_scores, best_model),
)
base_score = comp_scores.get(best_name)
final_params = best_params if optuna_score is None or base_score is None or optuna_score >= base_score else None
manual_metrics, sklearn_metrics = run_metrics_demo(data_split, split_key)
results = {
    "bagging": bagging_scores,
    "gradient_boosting": grad_scores,
    "comparison": comp_scores,
    "best_model": best_model,
    "optuna_best": best_name,
    "optuna_score": optuna_score,
    "final_params": final_params,
    "manual_metrics": manual_metrics,
    "sklearn_metrics": sklearn_metrics,
}
results
```

```python
run_submission(best_model_name=best_name, best_params=final_params)
Path(SUBMISSION_PATH).resolve()
```
